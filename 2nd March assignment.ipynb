Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Answer: Scatter plot uses dots to represent values for two different numeric variables.

```python
import numpy as np
import matplotlib.pyplot as py
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

py.scatter(x,y,c='r')
py.xlabel("x")
py.ylabel("y")
py.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Answer: 
subplot() function is used to drwa multiple plots in one figure.

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
py.subplot(2,2,1)
py.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
py.subplot(2,2,2)
py.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
py.subplot(2,2,3)
py.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
py.subplot(2,2,4)
py.plot(x,y)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Answer: 
Bar graph (bar chart) is a graph that represents the categorical data using rectangular bars. 

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

py.bar(company,profit)
py.xlabel("Company")
py.ylabel("Profit")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Answer:
A box plot is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum. The term "box plot" comes from the fact that the graph looks like a rectangle with lines extending from the top and bottom.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box_d=[box1,box2]

fig=py.figure(figsize =(10, 7))
ax=fig.add_axes([1,1,2,2])
bp=ax.boxplot(box_d)
py.show()
```
